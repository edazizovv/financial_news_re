```python
import os
import pandas
from matplotlib import pyplot
```

```python
#
```

```python
codes = ['no_news', 'with_news']
twos = ['./raw/121/', './raw/211/']
files = []
for j in range(len(twos)):
    for file in os.listdir(twos[j]):
        filek = pandas.read_excel((twos[j] + file))
        filek['estimator'] = file
        filek['report'] = codes[j]
        files.append(filek)
files = pandas.concat(files, axis=0, ignore_index=True)
files
```

```python
files.to_excel('./reported&muted/flat_report.xlsx', index=False)
```

```python
# Approach I: scoring by mean(nxt_fold, tst_fold)
```

```python
# we need aggregation by model X params X smoother !!
files['2mean_score'] = (files['R2_adj_nxt_fold'] + files['R2_adj_test']) / 2
files
```

```python
files[files['report'] == 'no_news'].sort_values(by=['2mean_score', 'Nf'], ascending=False)[
    ['Nf', '2mean_score', 'R2_adj_cur_fold', 'R2_adj_nxt_fold', 'R2_adj_test', 
'smoother', 'params', 'estimator']]
```

```python
files[files['report'] == 'with_news'].sort_values(by=['2mean_score'], ascending=False)[
    ['Nf', '2mean_score', 'R2_adj_cur_fold', 'R2_adj_nxt_fold', 'R2_adj_test', 
'smoother', 'params', 'estimator']]
```
