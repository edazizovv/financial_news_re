```python
start = datetime.datetime(year=2020, month=1, day=1, hour=0, minute=0, second=0, microsecond=0, tzinfo=pytz.timezone('EST'))
end = datetime.datetime(year=2020, month=5, day=1, hour=0, minute=0, second=0, microsecond=0, tzinfo=pytz.timezone('EST'))
test_range = pandas.DataFrame(data=pandas.date_range(start=start, end=end, freq='D'), columns=['base'])
test_range
```

```python
from busy_exchange.utils import BusyDayExchange
```

```python
test_range['is_holi'] = test_range['base'].apply(func=BusyDayExchange.is_holi)
test_range['is_full'] = test_range['base'].apply(func=BusyDayExchange.is_full)
test_range['is_cut'] = test_range['base'].apply(func=BusyDayExchange.is_cut)
test_range
```

```python
run_time = time.time()
test_range['to_holi'] = test_range['base'].apply(func=BusyDayExchange.to_holiday, args=(True,))
test_range['to_full'] = test_range['base'].apply(func=BusyDayExchange.to_fullday, args=(True,))
test_range['to_cut'] = test_range['base'].apply(func=BusyDayExchange.to_cutday, args=(True,))
test_range['af_holi'] = test_range['base'].apply(func=BusyDayExchange.to_holiday, args=(False,))
test_range['af_full'] = test_range['base'].apply(func=BusyDayExchange.to_fullday, args=(False,))
test_range['af_cut'] = test_range['base'].apply(func=BusyDayExchange.to_cutday, args=(False,))
run_time = time.time() - run_time
print(run_time)
test_range
```
