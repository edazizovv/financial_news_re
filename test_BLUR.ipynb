```python
import json
import time
import numpy
import pandas
import datetime
```

```python
from glaive_thrower import glaive_thrower
from supp import Neakt, OLR
```

```python
d = './result.csv'
data = pandas.read_csv(d)
# data = data.rename(columns={'lag': 'news_horizon'})
data = data.set_index(['ticker', 'time', 'news_horizon'], drop=False)
data = data.sort_index()
data
```

```python
y_name = 'open_LAG0'
removes = ['ticker', 'time', 'id', 'title', 'news_time'] + [x for x in data.columns.values if 'LAG0' in x and x != y_name]

news_columns = []
quotes_columns = [x for x in data.columns.values if 'LAG' in x and 'LAG0' not in x]
qualitative = []
quantitative = quotes_columns + ['news_horizon']

X_names = [quantitative + [y_name], quantitative]
Y_names = [y_name, y_name]
outputs_0 = quantitative + [y_name]
output_spec = [{x: 'float64' for x in outputs_0}, {y_name: 'float64'}]
```

```python
rfe_cv = False

report, plots = glaive_thrower(code='dead_dog', 
                               models=[OLR], transformators=[Neakt], 
                               data=data, y_name=y_name, removes=removes, 
                               news_columns=news_columns, quotes_columns=quotes_columns, 
                               qualitative=qualitative, quantitative=quantitative, 
                               X_names=X_names, Y_names=Y_names, output_spec=output_spec, 
                               rfe_cv=rfe_cv)

report
```
