```python
import numpy
from matplotlib import pyplot
```

```python
def ded_log(array):
    arr = array.copy()
    mask_plus = array > 0
    mask_minus = array < 0
    mask_zero = array == 0
    arr[mask_plus] = numpy.log(array[mask_plus])
    arr[mask_minus] = -numpy.log(-array[mask_minus])
    arr[mask_zero] = 0
    return arr
```

```python
def ded_exp(array, mask_plus, mask_minus, mask_zero):
    arr = array.copy()
    arr[mask_plus] = numpy.exp(array[mask_plus])
    arr[mask_minus] = -numpy.exp(-array[mask_minus])
    arr[mask_zero] = 0
    return arr
```
